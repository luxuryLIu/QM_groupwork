```python
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt

# lord data
file_path = "violentcrime_borough.geojson"
gdf = gpd.read_file(file_path)

# check if vail
gdf = gdf[gdf.geometry.notnull()]

# check data structure
print(gdf.head())
# print(gdf['year'])
```

```python
import matplotlib.pyplot as plt
# divide with year
df_2011 = gdf[gdf['year'] == 2011.0]
df_2021 = gdf[gdf['year'] == 2021.0]

# save as csv
df_2011.to_csv("borough_data_2011.csv", index=False)
df_2021.to_csv("borough_data_2021.csv", index=False)

print("'borough_data_2011.csv' and  'borough_data_2021.csv' saved")
```

```python
import matplotlib.pyplot as plt

# set graph size and font
plt.rcParams.update({'font.size': 12})

df_2021_sorted = df_2021.sort_values(by='rateper1000', ascending=False)

# Crime Rate per 1000 Population
plt.figure(figsize=(12, 6))
plt.bar(df_2021_sorted['name'], df_2021_sorted['rateper1000'], color='skyblue')
plt.xlabel('Borough')
plt.ylabel('Crime Rate per 1000 Population')
plt.title('Crime Rate per 1000 Population by Borough (2021)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```
